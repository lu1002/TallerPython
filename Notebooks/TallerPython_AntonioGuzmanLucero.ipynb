```python
cal = [9,7,9,9,8]

s = 0
for i in range(len(cal)):
  s += cal[i]
p = s / len(cal)
print(f'Promedio de calif: {p}')

s = 0
for i in cal:
  s += i
p = s / len(cal)
print(f'Promedio de calif: {p}')
```

Actividad 2: Dado 15 numeros, imprimir cantidad de pares y convertir los impares a pares

```python
nums = []
for i in range(15):
  nums.append(int(input(f'Valor {i+1}: ')))

par = 0
c = 0
for i in nums:
  if i % 2 == 0:
    par += 1
  else:
    nums[c] += 1
  c += 1

print(f'Cantidad de numeros pares: {par}')
print(f'Lista completa: {nums}')
```
